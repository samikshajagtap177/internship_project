```python
# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Sort values by Date
df = df.sort_values('Date')
```

```python
# Remove duplicates
df = df.drop_duplicates()
```

```python
# Reset index
df = df.reset_index(drop=True)
print("Data cleaned successfully!")
```

```python
# Create Daily Price Change Column
df['Price_Change'] = df['Close'] - df['Open']
```

```python
#Create Daily Return Percentage
df['Daily_Return_%'] = ((df['Close'] - df['Open']) / df['Open']) * 100
```

```python
#Create Moving Average (7 Days & 30 Days)
df['MA_7'] = df['Close'].rolling(window=7).mean()
df['MA_30'] = df['Close'].rolling(window=30).mean()
```

```python
#Line Chart 
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Close'])
plt.title("AAVE Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.xticks(rotation=45)
plt.show()
```

```python
#Line Chart 
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Volume'])
plt.title("AAVE Trading Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.xticks(rotation=45)
plt.show()
```

```python
#Bar Chart 
df['Month'] = df['Date'].dt.to_period('M')
monthly_avg = df.groupby('Month')['Close'].mean()

plt.figure(figsize=(10,5))
monthly_avg.plot(kind='bar')
plt.title("Monthly Average Closing Price")
plt.xlabel("Month")
plt.ylabel("Average Close Price")
plt.xticks(rotation=45)
plt.show()
```

```python
#Histogram 
plt.figure(figsize=(8,5))
plt.hist(df['Close'], bins=30)
plt.title("Distribution of Closing Prices")
plt.xlabel("Close Price")
plt.ylabel("Frequency")
plt.show()
```

```python
#Scatter Plot – Open vs Close
plt.figure(figsize=(6,5))
plt.scatter(df['Open'], df['Close'])
plt.title("Open vs Close Price")
plt.xlabel("Open Price")
plt.ylabel("Close Price")
plt.show()
```

```python
#Correlation
plt.figure(figsize=(8,6))
sns.heatmap(df[['High','Low','Open','Close','Volume','Marketcap']].corr(), 
            annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
#Market Cap Trend
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Marketcap'])
plt.title("Market Capitalization Over Time")
plt.xlabel("Date")
plt.ylabel("Market Cap")
plt.xticks(rotation=45)
plt.show()
```
